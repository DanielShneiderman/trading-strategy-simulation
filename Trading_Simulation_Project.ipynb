# Trading Strategy Monte Carlo Simulation

This project simulates a trading strategy using random outcomes.

### Metrics evaluated:
- Expected profit
- Drawdowns (risk)
- Profit distribution
- Drawdown distribution
- Risk-adjusted return

The goal is to analyze how randomness impacts trading performance and risk, including the distribution of profits and drawdowns.
This simulation assumes independent trades with fixed outcomes and probabilities.

```python
# Trading Strategy Monte Carlo Simulation.
#
# This project simulates a trading strategy using random outcomes.
# It evaluates performance using:
# - Expected profit
# - drawdowns (risk)
# - Profit distribution
# - Risk-adjusted return
#
# Assumes a fixed risk/reward profile (win = 10, loss = -8)
# To evaluate long-term expectancy under randomness.

import random
import statistics
import matplotlib.pyplot as plt

num_simulations = 100
num_trades = 1000
win_amount = 10
loss_amount = -8

results = []
drawdowns = []

for sim in range(num_simulations):
  total = 0
  peak = 0
  max_drawdown = 0

  for i in range(num_trades):
    trade = random.choice([win_amount, loss_amount])
    total += trade

    if total > peak:
      peak = total

    drawdown = peak - total

    if drawdown > max_drawdown:
      max_drawdown = drawdown

  results.append(total)
  drawdowns.append(max_drawdown)

average_profit = sum(results) / len(results)
min_profit = min(results)
max_profit = max(results)

average_drawdown = sum(drawdowns) / len(drawdowns)
worst_drawdown = max(drawdowns)
best_drawdown = min(drawdowns)

std_profit = statistics.stdev(results)
risk_adjusted_return = average_profit / std_profit


print("Average final profit:", average_profit)
print("Worst final profit:", min_profit)
print("Best final profit:", max_profit)
print("Average drawdown:", average_drawdown)
print("Worst drawdown:", worst_drawdown)
print("Best drawdown:", best_drawdown)
print("Profit standart deviation:", std_profit)
print("Risk-adjusted return:", risk_adjusted_return)


plt.hist(results, bins = 20)
plt.title("Distribution of Final Profit")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()

plt.hist(drawdowns, bins = 20)
plt.title("Distribution of Drawdowns")
plt.xlabel("Drawdown")
plt.ylabel("Frequency")
plt.show()





equity = []
total = 0

for i in range(num_trades):
  trade = random.choice([win_amount, loss_amount])
  total += trade
  equity.append(total)

plt.plot(equity)
plt.title("Equity Cruve")
plt.xlabel("Trade Number")
plt.ylabel("PnL")
plt.show()




```
